```python
try:
    marks = 10000
    a = marks / 0
    print(a)
except ZeroDivisionError:
    print("There is error: division by zero") 
    print("Pls check once")
else:
    print("no error in the code")
finally:
    print("done")
```

```python
try:
    x = 5
    y = "hello"
    z = x + y
except TypeError:
    print("There is error in the code: unsupported operand operation b/w string and int" )
    print("Pls check the code")
else:
    print("no error in the code")
finally:
    print("done")
```

```python
a = [1, 2, 3]
index = int(input(f"What index value do you want start with 0 to {len(a)-1} :"))
try:
    print (f"element value  {a[index]} at index {index}")
except IndexError:
    print("There is error in the code: list index out of range")
    print("Pls check once")
else:
    print("no error in the code")
finally:
    print("done")
```
